```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
```

```python
rows = 100
```

```python
customers = np.random.randint(20, 100, rows)
```

```python
hours = np.random.randint(4, 11, rows)
```

```python
drinks = customers * np.random.randint(2, 5, rows)
```

```python
day_type = np.random.choice(
    ["Weekday", "Weekend"],
    rows
)
```

```python
tips = (
    customers * 0.8 +
    drinks * 0.15 +
    np.where(day_type == "Weekend", 20, 0) +
    np.random.normal(0, 8, rows)
)
```

```python
df = pd.DataFrame({
    "Customers": customers,
    "Hours_Worked": hours,
    "Drinks_Sold": drinks,
    "Day_Type": day_type,
    "Tips": tips.round(2)
})
```

```python
df.to_csv("bartender_tips.csv", index=False)
```
